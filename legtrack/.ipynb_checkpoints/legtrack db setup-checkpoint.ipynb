```python
import sqlalchemy as db
from sqlalchemy import Integer, String
db_uri = 'sqlite:///lims_db.sqlite'
engine = db.create_engine(db_uri)

con = engine.connect()
metadata = db.MetaData()

# Declare a table
table = sqlalchemy.Table('bills',metadata,
              sqlalchemy.Column('ActNumber',String),
              sqlalchemy.Column('Category',String),
              sqlalchemy.Column('CommitteeReferral',String),
              sqlalchemy.Column('CommitteeReferralComments',String),
              sqlalchemy.Column('CouncilPeriod',String),
              sqlalchemy.Column('Deemed',String),
              sqlalchemy.Column('DeemedOn',String),
              sqlalchemy.Column('DocumentUrl',String),
              sqlalchemy.Column('Id',String),
              sqlalchemy.Column('IntroductionDate',String),
              sqlalchemy.Column('LawNumber',String),
              sqlalchemy.Column('LegislationNumber',String),
              sqlalchemy.Column('LegislationStatus',String),
              sqlalchemy.Column('Modified',String),
              sqlalchemy.Column('ShortTitle',String),
              sqlalchemy.Column('Subcategory',String)
             )

metadata.create_all(engine) #Creates the table
```

```python
from calendar import monthrange

        
for month_num in range(8):
   
    month_num +=1

    year = 2019

    month_length = monthrange(year, month_num)

    kwarg = {"StartDate": "{}/01/2019".format(str(month_num)),
             "EndDate": "{}/{}/2019".format(str(month_num),str(month_length[1])),
             "rowLimit":"10000"}
    
    print(kwarg)

    a = dclims.post.advancedSearch(**kwarg)
    data = pd.read_json(a.text)
    
    if month_num == 1:
        keys = data.keys()
        headers = []
        for key in keys:
            headers.append(key)
        with open('bills.csv','w') as f:
            for h in headers:
                f.write("," + h )
            f.write('\n')
    
    data.to_csv('bills.csv',mode='a', header=False)
```
